```python
import csv

data = [
    ["Region", "Country", "Product", "Sales", "Units", "Quarter", "Salesperson"],
    ["Asia", "India", "Laptop", 240000, 120, "Q1", "Raj"],
    ["Asia", "India", "Monitor", 95000, 95, "Q2", "Raj"],
    ["Asia", "Japan", "Monitor", 210000, 105, "Q1", "Yuki"],
    ["Asia", "Japan", "Laptop", 180000, 90, "Q3", "Yuki"],
    ["Asia", "Singapore", "Keyboard", 180000, 360, "Q1", "Mei"],
    ["Asia", "Singapore", "Mouse", 60000, 300, "Q2", "Mei"],
    ["Asia", "China", "Laptop", 310000, 155, "Q2", "Wei"],
    ["Asia", "China", "Monitor", 140000, 70, "Q3", "Wei"],
    ["Europe", "Germany", "Laptop", 110000, 55, "Q1", "Hans"],
    ["Europe", "Germany", "Keyboard", 45000, 90, "Q3", "Hans"],
    ["Europe", "France", "Mouse", 70000, 350, "Q2", "Claire"],
    ["Europe", "France", "Laptop", 130000, 65, "Q4", "Claire"],
    ["Europe", "Italy", "Keyboard", 60000, 120, "Q1", "Marco"],
    ["Europe", "Italy", "Monitor", 80000, 40, "Q4", "Marco"],
    ["Europe", "UK", "Laptop", 150000, 75, "Q2", "James"],
    ["Europe", "UK", "Mouse", 40000, 200, "Q3", "James"],
    ["North America", "USA", "Laptop", 320000, 160, "Q1", "John"],
    ["North America", "USA", "Monitor", 210000, 105, "Q2", "John"],
    ["North America", "Canada", "Keyboard", 95000, 190, "Q1", "Sarah"],
    ["North America", "Canada", "Mouse", 55000, 275, "Q3", "Sarah"],
    ["North America", "Mexico", "Laptop", 75000, 37, "Q2", "Carlos"],
    ["North America", "Mexico", "Keyboard", 40000, 80, "Q4", "Carlos"],
    ["Middle East", "UAE", "Monitor", 90000, 45, "Q1", "Ahmed"],
    ["Middle East", "UAE", "Laptop", 120000, 60, "Q3", "Ahmed"],
    ["Middle East", "Saudi Arabia", "Laptop", 85000, 42, "Q2", "Khalid"],
    ["Middle East", "Saudi Arabia", "Keyboard", 50000, 100, "Q4", "Khalid"],
    ["Middle East", "Qatar", "Monitor", 70000, 35, "Q1", "Fatima"],
    ["Middle East", "Qatar", "Mouse", 30000, 150, "Q3", "Fatima"],
]

with open("sales.csv", "w", newline="") as f:
    writer = csv.writer(f)
    writer.writerows(data)

print("sales.csv created!")
```

```python
from langchain_openai import ChatOpenAI
from langchain_experimental.agents import create_csv_agent

llm = ChatOpenAI(model="gpt-4o", temperature=0)

agent = create_csv_agent(
    llm,
    "sales.csv",
    verbose=True,
    allow_dangerous_code=True
)
```

```python
agent.invoke({"input": "Which salesperson has the lowest total sales and in which region do they work?"})
```

```python
agent.invoke({"input": "Who is the top salesperson in Asia, what product do they sell most, and what is their total sales in Q1?"})
```
